## Passing Learners from the last N days in self-paced non-differentiated courses

```python
days_since_completion = 14
days_before_ver_deadline = 5
```

```python
%%vertica (days_since_completion=days_since_completion, days_before_ver_deadline=days_before_ver_deadline, excluded_courses=excluded_courses)
DROP TABLE IF EXISTS experimental_user_population;
CREATE local TEMPORARY TABLE experimental_user_population
ON COMMIT PRESERVE ROWS AS
SELECT DISTINCT
    a.user_id,
    d.user_username,
    d.user_email,
    a.passed_timestamp,
    a.course_id,
    c.course_name,
    c.course_seat_price,
    c.course_verification_end_date,
    c.course_partner,
    e.name,
    f.sku,
    RANDOMINT(2) as segment
FROM business_intelligence.course_completion_user a
JOIN production.d_user_course b
    ON a.passed_timestamp is not null
    AND a.passed_timestamp >= (CURRENT_DATE - :days_since_completion)
    AND b.current_enrollment_mode IN ('honor','audit')
    AND b.current_enrollment_is_active = 1  -- added is_active check
    AND a.user_id = b.user_id
    AND a.course_id = b.course_id
JOIN production.d_user d
    ON d.user_email is not null
    AND a.user_id = d.user_id
JOIN business_intelligence.course_master c
    ON a.course_id = c.course_id
    AND c.course_verification_end_date >= (CURRENT_DATE + :days_before_ver_deadline)
    AND c.pacing_type = 'self_paced'
    AND c.content_language = 'en-us'
    AND c.is_WL = 0
JOIN lms_read_replica.auth_userprofile e
    ON a.user_id = e.user_id
JOIN lms_read_replica.course_modes_coursemode f
    ON f.course_id = a.course_id
    AND f.mode_slug = 'verified'
WHERE a.course_id not in :excluded_courses
```

```python
%%vertica -> deduped_experimental_user_population
SELECT 
*
FROM (
    select
        user_id,
        user_username,
        course_id,
        user_email,
        passed_timestamp,
        segment,
        course_name, 
        course_seat_price,
        course_verification_end_date,
        course_partner,
        name,
        sku,
        RANK() OVER(PARTITION BY user_id ORDER BY passed_timestamp DESC) as rank
    FROM
        experimental_user_population
    ORDER BY user_id, passed_timestamp DESC
) as ordered_population
where rank = 1
```
