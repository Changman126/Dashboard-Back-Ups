```python
bulk = pd.read_csv('../Documents/bulk_purchase.csv')
```

```python
bulk.groupby('user_id').aggregate('ActionType')
```

```python
grouped = bulk[bulk.Rank<=2].sort_values(['user_id','Rank']).groupby('user_id')

df = grouped.aggregate(lambda x: tuple(x)).reset_index()
df.groupby('ActionType').count().sort_values('user_id', ascending=False)/df.groupby('ActionType').count().sum()['user_id']
```
