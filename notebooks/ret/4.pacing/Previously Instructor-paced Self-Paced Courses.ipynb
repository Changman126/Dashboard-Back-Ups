```python
%%vertica -> formerly_instructor_self_paced
SELECT
 a.course_number,
 a.course_id AS course_id_self_paced_active,
 a.course_name AS course_name_self_paced_active,
 a.pacing_type AS pacing_type_self_paced_active,
 a.is_active AS is_active_self_paced_active,
 b.course_id AS course_id_instructor_paced_inactive,
 b.course_name AS course_name_instructor_paced_inactive,
 b.pacing_type AS pacing_type_instructor_paced_inactive,
 b.is_active AS is_active_instructor_paced_inactive
FROM 
 business_intelligence.course_master a 
JOIN 
 business_intelligence.course_master b
ON 
 a.course_number = b.course_number
AND 
 a.is_active
AND 
 NOT b.is_active
AND 
 a.pacing_type = 'self_paced'
AND 
 b.pacing_type = 'instructor_paced'
```
